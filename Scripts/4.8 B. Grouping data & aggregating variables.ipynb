```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

01. Importing data

```python
# Import data
path = r'C:\Users\polly\Documents\Instacart Basket Analysis'
df_ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_derived.pkl'))     
```

02. Grouping and aggregating data

```python
# Group data by department_id column and aggregate by order_number column for entire dataframe
df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

Since we are using the entire dataframe, the output shows all of the department id's (from 1 to 21) in the output, rather than just those included in the first million rows.

03. Creating customer loyalty flag

```python
# Create a new column 'max_order' that shows the maximum orders for each user
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check output
df_ords_prods_merge.head(15)
```

```python
# Create 'loyalty_flag' column based on max_order value for each user 
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Print frequency of 'loyalty_flag' column
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

04. Determine the spending habits of the three types of customers

```python
# Group data by 'loyalty_flag' column and aggregate by 'prices' column mean, min and max
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

05. Creating customer spending flag

```python
# Create a new column 'average_price' that shows the average price of items for user spent
df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output
df_ords_prods_merge.head(15)
```

```python
# Create a 'spending_flag' column based on the average_price value for each user 
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Print frequency of 'spending_flag' column
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check output of first 60 rows of 'user_id', 'spending_flag' and 'average_price' columns
df_ords_prods_merge[['user_id', 'spending_flag', 'average_price']].head(60)
```

06. Creating customer order frequency flag

```python
# Create a new column 'median_days_since_prior_order' that shows the median number of days since prior order for each user
df_ords_prods_merge['median_days_since_prior_order'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check output
df_ords_prods_merge.head(15)
```

```python
# Create an 'order_frequency_flag' column based on the median_days_since_prior_order value for each user 
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_prior_order'] > 10) & (df_ords_prods_merge['median_days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Print frequency of 'order_frequency_flag' column
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Check output of first 60 rows of 'user_id', 'order_frequency_flag' and 'median_days_since_prior_order' columns
df_ords_prods_merge[['user_id', 'order_frequency_flag', 'median_days_since_prior_order']].head(60)
```

```python
# Change outlier 'prices' values to NaNs
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# Check max value on 'prices' column
df_ords_prods_merge['prices'].max()
```

```python
# Export df_ords_prods_merge to 'Prepared Data' folder
df_ords_prods_merge.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_derived_more.pkl'))
```
