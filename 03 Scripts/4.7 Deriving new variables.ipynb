02. Create a new user-defined function

```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return np.nan
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check frequency of values in new column
df['price_range'].value_counts(dropna = False)
```

```python
# Determine the most expensive item in the subset
df['prices'].max()
```

04. Create new column with if-statement and for-loop

```python
# Check the frequency of 'orders_day_of_week' column
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create empty list 'result' and assign values using for-loop and if-statements
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Set new column 'busiest day' equal to 'result'
df_ords_prods_merge['busiest_day'] = result
```

```python
# Check frequency of new column
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
# Check output
df_ords_prods_merge.head()
```

05. Task 4.7

```python
# Create empty list 'result_new' and assign new values
result_new = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result_new.append("Busiest days")
  elif value == 1:
    result_new.append("Busiest days") 
  elif value == 4:
    result_new.append("Slowest days")
  elif value == 3:
    result_new.append("Slowest days")
  else:
    result_new.append("Regularly busy")
```

```python
result_new
```

```python
# Set new column 'busiest_day_new' equal to 'result_new'
df_ords_prods_merge['busiest_day_new'] = result_new
```

```python
# Check frequency of new column 'busiest days'
df_ords_prods_merge['busiest_day_new'].value_counts(dropna = False)
```

Busiest days is the total of days '0' and '1', the two busiest days; Slowest days is the total of days '3' and '4', the two slowest days; Regularly busy is the total of the remaining days: '2', '5', and '6.

```python
# Create empty list 'hours' and assign new values
hours = []

for value in df_ords_prods_merge["order_time_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
   hours.append("Most orders")
  elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1 or value == 0 or value == 6 or value == 23:
    hours.append("Fewest orders") 
  else:
    hours.append("Average orders")
```
