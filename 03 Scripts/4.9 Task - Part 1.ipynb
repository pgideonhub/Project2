03. Data wrangling

```python
# Rename columns
df_customers.rename(columns = {'First Name' : 'first_name', 'Surnam' : 'surname', 'Gender': 'gender', 'STATE' : 'state', 'Age' : 'age', 'n_dependants' : 'number_of_dependents'}, inplace = True)
```

```python
# Check output
df_customers.head()
```

```python
# Change data types for numeric columns to reduce memory usage
df_customers['user_id'] = df_customers['user_id'].astype('int32')
df_customers['age'] = df_customers['age'].astype('int8')
df_customers['number_of_dependents'] = df_customers['number_of_dependents'].astype('int8')
df_customers['income'] = df_customers['income'].astype('int32')
             
```

```python
# Check datatypes and memory usage
df_customers.info()
```

04. Checking data consistency

```python
# Check statistics of numeric columns
df_customers.describe()
```

Statistics seem in order except possibly maximum salary of 593,901. This is within the realm of possibility, and could be an outlier. But it's also possible that it's a mistake. Will keep this in mind for future analysis.

```python
# Check for mixed data types
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

```python
# Change datatype of 'first_name' column to string
df_customers['first_name'] = df_customers['first_name'].astype('str')
```

```python
# Confirm that there are no more mixed data types
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```
